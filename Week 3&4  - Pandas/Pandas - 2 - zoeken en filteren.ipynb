We openen weer het bestand *ufo_sightings.csv*.

````python
import pandas as pd
ufos = pd.read_csv('ufo_sightings.csv')
ufos.head(5)
````

Er zijn verschillende manieren om door je DataFrame heen te zoeken en subselecties te maken. Een veelgebruikte manier is via het attribuut ```loc```. Dit attribuut kun je tussen blokhaken slicen op verschillende manieren. Je kan bijvoorbeeld direct filteren op de index:

````python
ufos.loc[1200]
````

Of op meerdere indexen:

````python
ufos.loc[[1200, 50, 4000]]
````

Of op een range van indexen:

````python
ufos.loc[1200:1205]
````

Je kan ook (net zoals je dat gewend bent bij het slicen van lijsten) indexen weglaten voor of na de komma. Bijvoorbeeld als je de eerste 5 rijen wilt zien. Let wel op: hij laat alle indexen zien **tot en met** index 5. Dit is dus anders dan je gewend bent bij het slicen van lijsten. Het statement ```ufos.loc[:5]``` doet dus hetzelfde als ```ufos.head(6)```.

````python
ufos.loc[:5]
````

We laden nu weer ons oorspronkelijke bestand in:

````python
ufos = pd.read_csv('ufo_sightings.csv')
````

Je kan ook aangeven welke kolommen je wilt zien. Bijvoorbeeld zo:

````python
ufos.loc[1200:1205,'comments']
````

Of zo:

````python
ufos.loc[1203,['city','shape','duration']]
````

Een laatste voorbeeld: toon alle rijen, en alleen de kolommen 'datetime' t/m 'duration': 

````python
ufos.loc[:,:'duration']
````

Nu naar het interessante gedeelte. Stel: je wilt alle disk-shaped UFO's weten die in San Antonio zijn gezien. Om te beginnen doe je het volgende:

````python
(ufos.loc[:,'city']=='san antonio') & (ufos.loc[:,'shape']=='disk')

#Wanneer er geen specifieke rijen hoeven te worden geselecteerd, kan dit ook korter worden geschreven:

(ufos['city']=='san antonio') & (ufos['shape']=='disk')
````

Het resultaat van het vorige statement is een Series van boolean waarden. Het lijkt overal False te zijn, maar als je goed kijkt zie je bij index 25 een True staan. Er wordt dus voor iedere rij aangegeven of de waarde van de kolom *city* wel of niet geijk is aan 'san antonio' **EN** de waarde van de kolom *shape* gelijk is aan 'disk'. 

Dit soort statements is overigens erg gevoelig voor haakjes. Als je de haakjes rond de eerste helft van het statement weghaalt, levert dit een foutmelding op.

Vervolgens kan je deze Series weer gebruiken om te slicen via het loc-attribuut. Dit noem je *boolean masking*:

````python
# .loc notatie
#mask = (ufos.loc[:,'city']=='san antonio') & (ufos.loc[:,'shape']=='disk')
#ufos[mask]

# Kortere notatie
mask = (ufos['city']=='san antonio') & (ufos['shape']=='disk')
ufos[mask]

# Query methode
ufos.query("city == 'san antonio' and shape == 'disk'")

# Query methode via variabele
city = 'san antonio'
shape = 'disk'
ufos.query("city == @city and shape == @shape")
````

Nog een voorbeeld (zonder de tussenvariabele). Kijk of je het hele statement kunt doorgronden:

````python
#ufos.loc[(ufos.loc[:,'duration']>5000) & ((ufos.loc[:,'country']=='de') | (ufos.loc[:,'country']=='gb'))]

# Masking methode
ufos[(ufos['duration']>5000) & ((ufos['country']=='de') | (ufos['country']=='gb'))]

# Query methode
ufos.query("duration > 5000 and (country == 'de' or country == 'gb')")
````
