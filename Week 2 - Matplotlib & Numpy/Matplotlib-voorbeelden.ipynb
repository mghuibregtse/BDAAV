```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.arange(0, 10, 1)
y1 = np.arange(100, 200, 10)
plt.plot(x1, y1, label='Lijn 1')

x2 = np.arange(0, 10, 1)
y2 = np.arange(0, 300, 30)
plt.plot(x2, y2, label='Lijn 2')

plt.title('Titel')
plt.ylabel('Y-as')
plt.xlabel('X-as')

plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
plt.xlim(0.0, 4.5)
plt.show()
```

```python
x = np.random.normal(size = 1000)
plt.hist(x, bins=10)
plt.ylabel('Aantal')
plt.show()
```

```python
x = np.array([0,1,2])
y = np.array([10,8,9])
plt.bar(x, y)
plt.xticks(x, ['a','b','c'])
plt.ylabel('Aantal')
plt.xlabel('Categorie')
plt.show()
```

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%.0f %%', startangle=90)
plt.show()
```
